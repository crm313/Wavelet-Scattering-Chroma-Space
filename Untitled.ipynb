```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


diffs = pd.read_csv('diffs.csv')
#subjects = ['Subject A (10)', 'Subject C (10)', 'Subject B (8)', 'Subject D (4)']
#sigma = [0.2341, 0.2450, 0.7710, 0.8028]
sns.set_context("paper", font_scale=1.5)
plt.figure(figsize=(8, 8))
ax = sns.stripplot(x="difference", y="Subject", data=diffs, palette="Blues_d")
ax.set(ylabel='Subject (Number Of Trials)', xlabel = "$\sigma$ (Centimeters)", title= "Interaural Distance Standard Deviation - MARL")
#show the plot
sns.plt.show()

```
