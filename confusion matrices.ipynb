```python
from pylab import *
import numpy as np
%matplotlib inline
chord_list = [ 'maj', 'min', 'min7', '7', 'N', 'maj7', 'sus4', 'maj6', 'min6', 'sus2', 'dim', 'aug', 'hdim7', 'dim7']
```

```python
mb4 = np.load('confusion-multiband-4.npy')
# rows: estimation
# columns: reference

# normalize
i = np.arange(14)
for ref in i:
    mb4[:,ref] = mb4[:,ref]/np.sum(mb4[:,ref])
figure(1)
plt.figure(figsize=(8,8))
imshow(mb4,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
plt.xlabel('Annotated Quality')
plt.ylabel('Estimated Quality')
#plt.title('Confusions - Multiband')
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
#ax.title.set_label_position('bottom') 
plt.savefig('confusion-multiband.png', format='png', dpi=100)
```

```python
w4 = np.load('confusion-wavelet-4.npy')

# normalize
i = np.arange(14)
for ref in i:
    w4[:,ref] = w4[:,ref]/np.sum(w4[:,ref])
figure(1)
plt.figure(figsize=(8,8))
imshow(w4,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
plt.xlabel('Annotated Quality')
plt.ylabel('Estimated Quality')
ax.xaxis.set_ticks_position('top') # the rest is the same
ax.xaxis.set_label_position('top') 
plt.savefig('confusion-wavelet.png', format='png', dpi=100)
```

```python
s4 = np.load('confusion-scattering-4.npy')

# normalize
i = np.arange(14)
for ref in i:
    s4[:,ref] = s4[:,ref]/np.sum(s4[:,ref])
figure(1)
plt.figure(figsize=(8,8))
imshow(s4,interpolation='none')
plt.xticks(np.arange(14))
plt.xlabel('Annotated Quality')
plt.yticks(np.arange(14))
plt.ylabel('Estimated Quality')
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
ax.xaxis.set_ticks_position('top') # the rest is the same
ax.xaxis.set_label_position('top')
plt.savefig('confusion-scattering.png', format='png', dpi=100)
```

```python
mb2w = mb4 - w4
mb2s = mb4 - s4
figure(1)
plt.figure(figsize=(12,12))
imshow(mb2w,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
ax.xaxis.set_ticks_position('top') # the rest is the same

figure(2)
plt.figure(figsize=(12,12))
imshow(mb2s,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
ax.xaxis.set_ticks_position('top') # the rest is the same
```

```python
w2m = w4 - mb4
w2m.clip(min=0)
s2m = s4 - mb4
s2m.clip(min=0)

figure(1)
plt.figure(figsize=(12,12))
imshow(w2m,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
ax.xaxis.set_ticks_position('top') # the rest is the same

figure(2)
plt.figure(figsize=(12,12))
imshow(s2m,interpolation='none')
plt.xticks(np.arange(14))
plt.yticks(np.arange(14))
ax = plt.gca()
ax.set_xticklabels(chord_list)
ax.set_yticklabels(chord_list)
ax.xaxis.set_ticks_position('top') # the rest is the same
```
